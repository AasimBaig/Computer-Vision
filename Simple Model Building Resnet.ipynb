```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
## importing resenet50
from keras.applications import ResNet50

## resnet50 preprocess input contain image preprocessing before feeding it to network 
## decode_prediction -> decodes our image prediction
from keras.applications.resnet50 import preprocess_input , decode_predictions

## load_img -> loads our image
## img_to_array -> convert our image to array
from keras.preprocessing.image import load_img,img_to_array
```

```python
## lets take our 3 image of doggies <3
## bulldog, siberian_huskey, pug
## we can write function that loads the image from given directory but we have 3 images only
img_path = ["images/bulldog.jpg",
           "images/siberian_huskey.jpg",
           "images/pug.jpg"]


## Resnet 50 takes 224 height and width of image so,
image_size = 224
```

```python
img_height = image_size
img_width = image_size
bulldog = load_img(img_path[0],target_size=(img_height, img_width))
sh = load_img(img_path[1],target_size=(img_height, img_width))
pug = load_img(img_path[2],target_size=(img_height, img_width))
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(15,15))
ax[0].imshow(bulldog)
ax[0].set_title("Bulldog")
ax[1].imshow(sh)
ax[1].set_title("Siberian Huskey")
ax[2].imshow(pug)
ax[2].set_title("Pug")
plt.plot()
```

```python
## lets load our images, convert to array and preprocess our image
def read_and_prep_images(img_paths, img_height=image_size, img_width=image_size):
    imgs = [load_img(img_path, target_size=(img_height, img_width)) for img_path in img_paths]
    img_array = np.array([img_to_array(img) for img in imgs])
    output = preprocess_input(img_array)
    return(output)
```

```python
# preprocessing our images
img = read_and_prep_images(img_path)
```

```python
# lets load resnet50 model using model function
model = ResNet50(weights="imagenet")
```

```python
# model.predicts function is use to predict the image
preds = model.predict(img)
```

```python
# here we decode our predictions
most_likely = decode_predictions(preds)
```

```python
print(most_likely[0])
print(most_likely[1])
print(most_likely[2])
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(15,15))
ax[0].imshow(bulldog)
ax[0].set_title(most_likely[0][0])
ax[1].imshow(sh)
ax[1].set_title(most_likely[1][0])
ax[2].imshow(pug)
ax[2].set_title(most_likely[2][0])
plt.plot()
```
